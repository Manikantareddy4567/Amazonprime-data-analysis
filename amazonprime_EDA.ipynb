```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
prime_dataset=pd.read_csv(r"C:\prime-data-analysis\amazon_prime_titles.csv")
```

```python
prime_dataset.head(10)
```

```python
prime_dataset.shape
```

```python
prime_dataset.isnull().sum()
```

```python
prime_dataset[prime_dataset["date_added"].isnull()==True]
```

```python
prime_dataset.drop(columns=['date_added','country'],inplace=True)
```

```python
prime_dataset['director'].fillna('Unknown',inplace=True)
prime_dataset['cast'].fillna('Unknown',inplace=True)
prime_dataset['rating'].fillna(prime_dataset['rating'].mode()[0],inplace=True)
```

```python
prime_dataset.head(10)
```

```python
prime_dataset.isnull().sum()
```

```python
prime_dataset.duplicated().sum()
```

```python
prime_dataset['type'].unique()
```

```python
df=prime_dataset.copy()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='type', order=df['type'].value_counts().index)
plt.title("Distribution of Content Type")
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.histplot(df['release_year'], bins=30, kde=False)
plt.title("Number of Titles by Release Year")
plt.show()
```

```python
plt.figure(figsize=(14, 6))
df['rating'].value_counts().head(10).plot(kind='bar', color='orange')
plt.title("Top 10 Ratings")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(14, 6))
df['director'].value_counts().head(10).plot(kind='bar', color='green')
plt.title("Top 10 Directors")
plt.ylabel("Number of Titles")
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.countplot(data=df, x='release_year', hue='type', order=sorted(df['release_year'].unique()))
plt.xticks(rotation=90)
plt.show()
```

```python
from collections import Counter
genres = df['listed_in'].dropna().str.split(', ')
flat_genres = [genre for sublist in genres for genre in sublist]
pd.Series(Counter(flat_genres)).sort_values(ascending=False).head(10).plot(kind='barh')
```

```python
cast_series = df['cast'].dropna().str.split(', ')
actor_counts = Counter([actor for sublist in cast_series for actor in sublist])
pd.Series(actor_counts).sort_values(ascending=False).head(10).plot(kind='bar')
```
